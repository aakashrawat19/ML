```python
import pandas as pd
reviews = pd.read_csv("/home/luffy/Documents/winemag-data_first150k.csv", index_col=0)
reviews.head(10)
```

```python
reviews['points'].value_counts().plot.bar()        ##### here we can see x-axis values are not sorted
```

```python
reviews['points'].value_counts().sort_index().plot.bar()     ###### here we have sorted by index plot
```

```python
reviews['points'].value_counts().sort_index().plot.bar(figsize=(12,6))     ###### fig size is used to increase the height and
                                                                            ##### width of the graph
```

```python


reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='mediumvioletred'
)

#####to change the color of the graph and this is "midum voilet red"
```

```python
reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='red',
    fontsize=16
)


###### font size has increased to 16
```

```python
#### pandas graphics tool is buit up on matplotlib
reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='blue',
    fontsize=16,
    title='unbelivable'
)
```

```python
######### same can pe done  using matpliotlib 
import matplotlib.pyplot as plt
reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='red',
    fontsize=16
)
plt.title('Unexpected',fontsize=20)
```

```python

import seaborn as sns

ax = reviews['points'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16
)
ax.set_title("Rankings Given by Wine Magazine", fontsize=20)
sns.despine(bottom=True, left=True)
```
