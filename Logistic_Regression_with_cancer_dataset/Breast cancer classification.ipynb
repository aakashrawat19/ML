```python
import pandas as pd   ######### pandas for data manipulation ,dataframe ,read_csv
```

```python
import numpy as np   ######### matrix operation
from sklearn.metrics import accuracy_score ######### accuracy of the model
from sklearn.metrics import confusion_matrix  ###### confusion matrix
```

```python
from sklearn.linear_model import LogisticRegression   ############ OUR model
from sklearn.preprocessing import LabelEncoder    ######### used for converting categorical values into numerical values
from sklearn.model_selection import train_test_split  ######## splitting of dataset
```

```python
cancer = pd.read_csv('cancer/data.csv')  ######## reading data from disk
```

```python
cancer.head()   
```

```python
cancer.shape    ########## shape of the dataset
```

```python
cancer = cancer.drop(['id','Unnamed: 32'],axis = 1)    ######### droping 2 columns from our dataset ,namely id and unnamed32
```

```python
cancer['diagnosis'].unique()                  ################ finding unique values  of target value, M stands for malignant and B stands for Binign tumor
```

```python
cancer.describe()    ####### description of the dataset
```

```python
def encoding(cancer):    ####### function to encode categorical values to numerical values
    
    le = LabelEncoder()
    for col in cancer.columns:
        if cancer[col].dtypes =='object':            ####### when column's data type is equal to object
            cancer[col] = le.fit_transform(cancer[col]) 
    return cancer
            
def splitting(X,Y):    ######### function to split dataset in to training and testing dataset
    return train_test_split(X,Y,test_size =0.30,random_state = 42)  ######  dividing dataset
```

```python
encoded_cancer = encoding(cancer)
encoded_cancer.head()
```

```python
Y = cancer.diagnosis   ############ target varible
Y
```

```python
X = cancer.drop('diagnosis',axis = 1)   ############ input features
X.head()
```

```python
x_train,x_test,y_train,y_test = splitting(X,Y)  
```

```python
lr = LogisticRegression()          
lr.fit(x_train,y_train)           ######### fitting of data
y_pred = lr.predict(x_test)       #########  prediction
acc1 = accuracy_score(y_test,y_pred) ########## accuracy score
mat1 = confusion_matrix(y_pred,y_test)  ######### accuracy matrix
print(acc1)
print(mat1)
```

```python
from sklearn.feature_selection import SelectKBest  ######### selecting k best feaures from dataset for creating models
from sklearn.feature_selection import chi2         ######### ki square for feature selection
X_new = SelectKBest(chi2,15).fit_transform(X,Y)     
print(X_new.shape)
```

```python
x_train,x_test,y_train,y_test = splitting(X_new,Y)
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
acc1 = accuracy_score(y_test,y_pred)
mat1 = confusion_matrix(y_pred,y_test)
print(acc1)
print(mat1)
```

```python
############### Now comes the scartch part ########

n = len(Y)    ############ length of the dataset
X1 = np.ones((n,1))   ######  creating vector of ones 
X_transformed = pd.DataFrame(np.hstack((X1,X_new))  )      ############# appending one's in input feature
x_train,x_test,y_train,y_test = splitting(X_transformed,Y)  
```

```python
def linear_regression(x_train,y_train,x_test):  ######### function for linear regression
     

    x_trans = np.transpose(x_train)             ##### transpose
    product = np.dot(x_trans,x_train)           ####### multiplication 
    inverse = np.linalg.inv(product)            ###### inverse 
    product = np.dot(inverse,x_trans) 
    result = np.dot(product,y_train) 
    y_pred2 = np.dot(x_test,result) ######  prdiction using y_pred = x*b_pred
    return y_pred2
```

```python
def sigmoid(data):        ############ sigmoid funtion for logistic regression
    for i in range(len(data)):
        data[i] = 1/(1 + np.exp(-data[i]))
    return data
```

```python
y_pred2 = linear_regression(x_train,y_train,x_test)
y_pred3 = sigmoid(y_pred2)
y_pred3
for i in range(len(y_pred3)):    ################ this loop is to categorized the sample using thresold of 0.5
    if y_pred3[i] > 0.5 :
        y_pred3[i] = 1
    else:
        y_pred3[i] = 0
        
print(y_pred3)
```

```python
acc2 = accuracy_score(y_pred3,y_test)   

mat2 = confusion_matrix(y_pred3,y_test)
```
