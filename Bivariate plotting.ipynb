take limited amount of data for scatter plot because too many data can overlap each other and evntually you can lose the forest of the tree. here i take 150 instances  which  is nice enough


**Hexplot**

A hexplot aggregates points in space into hexagons, and then colorize those hexagons:

```
df.plot.hexbin(x='sepalL',y='petalL',gridsize =15) #### hex plot bwetween sepal length and petal length
```

Hex plot is showing clustering of petal length at value 1.5 

```
df.plot.bar(stacked=True)                           ##### Stacked bar plot
```

```
df.plot.area()                               #### Area plot
```


    Q1:- A scatter plot or hex plot is good for what two types of data?
    Q2:- What type of data makes sense to show in a stacked bar chart, but         not in a bivariate line chart?
     Q3:- What type of data makes sense to show in a bivariate line chart, but not in a stacked bar chart?
   Q4:-  Suppose we create a scatter plot but find that due to the large number of points it's hard to interpret. What are two things we can do to fix this issue?




   Ans 1:- Scatter plots and hex plots work best with a mixture of ordinal categorical and interval data.

   Ans 2:- Nominal categorical data makes sense in a (stacked) bar chart, but not in a (bivariate) line chart.

   Ans 3:- Interval data makes sense in a bivariate line chart, but not in a stacked bar chart.

  Ans 4:-  One way to fix this issue would be to sample the points. Another way to fix it would be to use a hex plot.
